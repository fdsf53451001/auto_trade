```python
# binance api : https://binance-docs.github.io/apidocs/spot/cn/#k

import requests
import matplotlib.pyplot as plt
# from dataprep.eda import *
import pandas as pd
import time # 引入time
import numpy as np
```

```python
df_base = pd.read_csv('BTCUSDT.csv')
df_compare = pd.read_csv('BNBUSDT.csv')
```

```python
base_close = list(df_base['1'])
compare_close = list(df_compare['1'])
print(np.corrcoef(base_close[:-1],compare_close[1:])[0][1])
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(base_close)
ax2.plot(compare_close)
fig.show()
```
